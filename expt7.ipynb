```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
wine = load_wine()
x = wine.data
y = wine.target
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled, y, test_size = 0.3, random_state = 42, stratify = y)
```

```python
#Logistic Regression without PCA

clf_no_pca = LogisticRegression(max_iter = 500)
clf_no_pca.fit(x_train,y_train)
y_pred_no_pca = clf_no_pca.predict(x_test)
acc_no_pca = accuracy_score(y_test, y_pred_no_pca)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
print("Without PCA")
print("Accuracy Score:",acc_no_pca)
matrix = confusion_matrix(y_test, y_pred_no_pca)
print("Confusion Matrix")
print(matrix)
print("Confusion Matrix with Visualization")
display = ConfusionMatrixDisplay(confusion_matrix = matrix)
display.plot()
plt.title("Confusion Matrix for Logistic Regression without PCA")
plt.show()

```

```python
#Logistic Regression with PCA

pca = PCA(n_components = 2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)

clf_pca = LogisticRegression(max_iter = 500)
clf_pca.fit(x_train_pca, y_train)
y_pred_pca = clf_pca.predict(x_test_pca)
acc_pca = accuracy_score(y_test,y_pred_pca)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
print("With PCA")
print("Accuracy Score:",acc_pca)
matrix = confusion_matrix(y_test, y_pred_pca)
print("Confusion Matrix")
print(matrix)
print("Confusion Matrix with Visualization")
display = ConfusionMatrixDisplay(confusion_matrix = matrix)
display.plot()
plt.title("Confusion Matrix for Logistic Regression with PCA")
plt.show()
```
